```python
def determine_modifier(score):
    return (score - 10)//2
```

```python
def generate_ability_score(diemin=1, diemax=6, n_dice=4, drop_lowest=True):
    """
    diemin: an integer; defines the lowest number on the die.
    diemax: an integer; defines the highest number on the die.
    n_dice: how many of said dice you're rolling.
    drop_lowest: determines whether or not you're dropping the lowest die roll in every roll. Defaulted to True.
    """
    result_dict={}
    for i in range(n_dice):
        result_dict['roll.{0}'.format(i)]= random.randint(diemin, diemax)
    big_total = sum(result_dict.values())
    drop_lowest_total = big_total - min(result_dict.values())
    if drop_lowest==False:
        return big_total
    else:
        return drop_lowest_total
```

```python
generate_ability_score(drop_lowest = True)
```

```python
def assign_ability_scores(diemin=1, diemax=6, n_dice=4, drop_lowest=True):
    """
    diemin: an integer; defines the lowest number on the die.
    diemax: an integer; defines the highest number on the die.
    n_dice: how many of said dice you're rolling.
    drop_lowest: determines whether or not you're dropping the lowest die roll in every roll. Defaulted to True.
    """
    ability_dict= {'Strength':0,'Dexterity':0,'Constitution':0,'Intelligence':0,'Wisdom':0,'Charisma':0}
    for i in ability_dict:
        ability_dict['{}'.format(i)] = generate_ability_score(diemin, diemax, n_dice, drop_lowest)
    return ability_dict
```

```python
assign_ability_scores()
```

```python
def assign_ability_modifiers(ability_dict):
    """
    ability_dict = dictionary generated by assign_ability_scores.
    """
    modifiers_dict = {'Strength':0, 'Dexterity':0, 'Constitution':0, 'Intelligence':0, 'Wisdom':0, 'Charisma':0}
    for i in ability_dict:
        modifiers_dict['{}'.format(i)] = determine_modifier(ability_dict[i])
    return modifiers_dict
```

```python
def assign_base_attack_bonus(progression_type, level):
    """
    Assigns a Base Attack Bonus (BAB) for the first attack of a character to an integer.
    
    ~~~~~~ARGS~~~~~~~
    progression_type: a string that defines how quickly base attack bonus progresses per level. 
        'fighter' means, as per fighter class in 3.5e manual, BAB starts at 1 and rises by 1 per level.
        'rogue' means, as per rogue class in 3.5e manual, BAB starts at 0 and rises by 1 per level.
                KNOWN BUG: actual rogue BAB progression pauses every 4 levels. That is not the case in this function.
        'caster' means, as per wizard class in 3.5e manual, BAB starts at 0 and rises by 1 per two levels.
    level: character's current level. Must be at least 1.
    """
    if level <= 0:
        return 'Level Error: level argument must be greater than 0'
    elif progression_type == 'fighter':
        base_attack_bonus = 0
        for lev in range(level):
            base_attack_bonus += 1
    elif progression_type == 'rogue':
        base_attack_bonus = -1
        for lev in range(level)[0:4]:
            base_attack_bonus += 1
            #for lev in range(level)[4]:
                #base_attack_bonus +=0
    elif progression_type == 'caster':
        base_attack_bonus = 0
        for lev in range(level//2):
            base_attack_bonus += 1
    return base_attack_bonus
```
